```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
#read the file and rename the columns
df  = pd.read_csv('datasets/2012_SAT_Results.csv')
df = df.set_axis(['dbn', 'school_name', 'num_test_takers', 'reading_avg_score', 'math_avg_score', 'writing_avg_score'], axis=1)
df
```

```python
#replace the 's' with NA
df.replace('s', np.nan, inplace=True)
df
```

```python
#check for the null
missign_value = df.isnull().sum()
missign_value
```

```python
#check for duplicate value
df.duplicated().sum()
```

```python
#prints the name of the column, and then prints the frequency count of each unique value in that column
for col in df.columns.values.tolist():
    print(col)
    print(df[col].value_counts())
```

```python
#check the missing value and count the missing value of each column
check_missing = df.isnull()
for col in check_missing.columns.tolist():
    print(col)
    print(check_missing[col].value_counts())
```

```python
#drop the missing value
df_clean = df.dropna()
df_clean
```

```python
#check the missing value and count the missing value of each column
check_missing = df_clean.isnull()
for col in check_missing.columns.tolist():
    print(col)
    print(check_missing[col].value_counts())
```

```python
#get the average of the reading score
reading_avg = df_clean['reading_avg_score'].astype('Float64').mean()
reading_avg
```

```python
#get the average of the writing score
writing_avg = df_clean['writing_avg_score'].astype('Float32').mean()
writing_avg
```

```python
#get the the average of the math score
math_avg = df_clean['math_avg_score'].astype('Float32').mean()
math_avg
```

```python
#Visualize the result
subjects = ['Reading', 'Math', 'Writing']
average = [reading_avg, math_avg, writing_avg]

plt.pie(average, 
        labels=subjects,
        startangle=90,
        shadow=True,
        explode=(0,0.1,0),
        autopct='%1.1f%%')

plt.title('Scholastic Aptitute Test')
plt.show()
```
